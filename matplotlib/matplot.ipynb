```python

a=np.array([1,2,3,4,5,6,7,8])
b=np.array([8,7,6,5,8,7,6,5])
plt.plot(a,b)
plt.plot(a,b,'o:b')  #for dotted line
```

for changing color

```python
a=np.array([1,2,3,4,5,6,7,8])
b=np.array([8,7,6,5,8,7,6,5])
plt.plot(a,b,color='g',ms=30,marker='*',mfc='r') # ms for size ,mfc for inner color,mec for outer color of star
```

# for dotted line 

```python
plt.plot(a,b,color='g',ms=30,marker='*',mfc='r',ls='dotted')
```
