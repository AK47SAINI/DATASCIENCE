```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv("Used_Bikes.csv")
plt.scatter(data['price'],data['brand'])
plt.show()
```

```python
data=pd.read_csv("company_sales_data.csv")
plt.plot(data)
plt.show()
```

```python
plt.plot(data['month_number'],data['facecream'],marker='o',linewidth=.9,label="face cream")
plt.plot(data['month_number'],data['facewash'],marker='o',linewidth=.9,label='facewash')
plt.plot(data['month_number'],data['shampoo'],marker='o',linewidth=.9,label='shampoo')
plt.plot(data['month_number'],data['toothpaste'],marker='o',linewidth=.9,label='toothpaste')

plt.plot(data['month_number'],data['moisturizer'],marker='o',linewidth=.9,label='moisturizer')

plt.plot(data['month_number'],data["bathingsoap"],marker='o',linewidth=.9,label='bathingsoap')
plt.ylabel("sales units in number")
plt.xlabel("month number")

plt.legend()
plt.show()
```

```python
plt.scatter(data['month_number'],data['toothpaste'],label="Tooth paste Sales data")
plt.grid(linestyle='--')
plt.xlabel('Month Number')
plt.ylabel('Number of units sold')
plt.title("Tooth paste Sales data")
plt.legend()

plt.show()
```

<h1> FOR PLOTTING THE MULTIPLE BARS OF DIFFERENT VALUES IN THE SAME BAR CHART <H1>

```python
width=.5
month=np.arange(len(data['month_number']))
plt.bar(month,data['facecream'],width=.4,color='g',label='facecream')
plt.bar(width+month,data['facewash'],width=.5,color='r',label='facewash')

plt.legend()
plt.show()
```

```python
width=.5
month=np.arange(len(data['month_number']))
plt.bar(month,data['facecream'],width=.5,color='g',label='facecream')
plt.bar(width+month,data['facewash'],width=.5,color='r',label='facewash')


plt.legend()
plt.savefig('doublebar.jpg')  #IT ALWAYS SHOULD BE CALLED BEFORE THE SHOW()
#IF WE CALL IT AFTER THE SHOW TH

plt.show()


#FOR SAVING THE GRAPH WE HAVE TO USE THE savefig('graphname.png')
```

```python
width=.5
month=np.arange(len(data['month_number']))
plt.bar(month+1,data['facecream'],width=.5,color='g',label='facecream')
plt.bar(width+month+1,data['facewash'],width=.5,color='r',label='facewash')

plt.legend()
plt.show()
```

```python
x=[1,2,3,4]

y1=[10,20,25,30]
y2=[30,25,20,10]
line1,=plt.plot(x,y1,label='line1')
line2,=plt.plot(x,y2,label='line2')
first_legend=plt.legend(handles=[line1],loc='upper left')
ax=plt.gca().add_artist(first_legend) #add first legend manually
plt.legend(handles=[line2],loc='upper right')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("multiple legends in a plot")
plt.show()
```
