```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
a=np.array([2,3,5,4,2])
b=np.array([1,5,7,3,2])          # ls='--' dashed line ls=linestyle              thickness of the line
plt.plot(a,b,marker='*',linestyle='dotted',mfc='green',mec='red',color='r',ms=20,linewidth=10) #ms stands for the marker size
#for dotted line we can use 'dotted',':' 
#for dashed line we can use 'dashed","--"
#for user defined then linestyle='-.'
plt.show() #for removing the above name of the graph
```

<h1> PLOTTING MULTIPLE GRAPHS IN THE SINGLE GRAPH </h1>

```python
a=np.array([2,4,4])
b=np.array([2,6,1])
# plt.title("relation between x and y")
# plt.xlabel("x axis")
# plt.xlabel("y axis")
plt.subplot(2,2,1)
plt.title("relation between a and b")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(a,b)

plt.subplot(2,2,2)
plt.title("relation between b and a")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(b,a)

#FOR GIVING THE TITLE TO THE WHOLE PLOTTED GRAPH
plt.suptitle("Relation")
```

```python
#TO GIVE THE RANDOM COLOR TO ALL THE POINTS WE USE THE FOLLOWING ATTRIBUTES

plt.title("relation between a and b") 
a=[8,5,3]

b=[3,2,5]
colors=["red",'green','blue']
sizes=[100,322,200]              
        #alpha for the brightness of the point varies from 0 to 1
plt.scatter(a,b,color=colors,s=sizes,marker='o',alpha=.3,cmap="viridis") #for the size we use s attribute
plt.colorbar() #to printing the color bar on the right side of the graph
```

<h1> Pie Charts</h1>


```python
a=[2,6,8,9]
label=["apple","banana",'orange','mango']
plt.pie(a,labels=label,startangle=90)
plt.show()
```

```python
#EXPLODE ATTRIBUTE IN THE PIE CHARTS

import matplotlib.pyplot as plt
a=[3,6,2,7]

label=["apple","banana",'orange','mango']
myexplode=[.2,.3,.3,.7]
mycolor=['g','b','r','y']
plt.pie(a,labels=label,explode=myexplode,shadow=True,colors=mycolor)
plt.legend(title='fruits',loc='upper right') #for showing the info of the color on the piechart
#loc for changing the position of the legend of the piechart
plt.show()
```
