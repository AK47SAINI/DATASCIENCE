```python
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np
a=np.array([12,20,30,45,67])


b=np.array([15,20,34,47,56])
plt.plot(a,b)
plt.xlabel("x-axis")  
plt.ylabel('y-axis')
plt.title('realtion between a and b')
plt.xticks(a,[15,30,45,60,75])   #for changing the points at the x axis
plt.yticks(b,[20,30,40,50,60])  #for changing the points at the y axis
plt.xlim(10,80)  #to limiting the values of the points at the x axis

plt.ylim(15,70) #to limiting the values of the points at the x axis


plt.show()
```
