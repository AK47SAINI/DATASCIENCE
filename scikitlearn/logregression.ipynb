```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



data=pd.read_csv('cust_satisfaction.csv')
# print(data.count())
# print(data.duplicated())
print(data.nunique())  #for checking the unique values of the attributes in the dataframe data

```

```python
print(data['Customer Type'].value_counts())
data.info()
```

```python
# print(data.isnull().sum())


# df=data.shape[0]-data.isnull()  #for removing the null values from the dataset
data.dropna(inplace=True)   #also can be done by this function dropna
print(data.isnull().sum())
```

```python
print(data.duplicated().sum())
data.drop_duplicates(inplace=True)  #for removing the duplicated data

print(data.duplicated().sum())
```

```python
loyal_customer=data[data['Customer Type']=='Loyal Customer']  #storing the loyal_customer into the another data frame
disloyal_customer=data[data['Customer Type']=='disloyal Customer']  #storing the disloyal customer in the another data frame

# print(loyal_customer.count())
print(loyal_customer.head())
# print(disloyal_customer.count())
print(disloyal_customer.head())
```

```python
loyal_customer1=loyal_customer.sample(21000)
print(loyal_customer1.count())
loyal_customer1.shape
```

```python
balance_df=pd.concat([loyal_customer1,disloyal_customer],axis=0)
# print(balance_df.count())
print(balance_df.head())
```

```python
cat_col=balance_df.select_dtypes(include='O')
num_col=balance_df.select_dtypes(exclude='O')
print(num_col.head())
print(cat_col.head())
print(num_col.shape)
print(cat_col.shape)
```

```python
cat_col
```

```python
onehot_data=encoder.fit_transform(cat_col).toarray()
print(onehot_data.shape)
print(onehot_data)
print
```

```python
one_hot=pd.DataFrame(onehot_data,columns=column_names)

# print(one_hot.isnull().sum())
```

```python
final_df=pd.concat([one_hot,num_col],axis=1)
print(final_df)

```
