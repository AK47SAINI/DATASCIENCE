```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

data=pd.read_csv('Position_Salaries.csv')
print(data.count())
```

```python
# print(data.head())
# cat_col=data.drop(['Level','Salary'],axis=1)
# print(cat_col)
num_col=data.drop(['Position'],axis=1)


posdt={'Business Analyst':0,'Junior Consultant':1,'Senior Consultant':2,'Manager':3,'Country Manager':4,'Region Manager':5,'Partner':6,'Senior Partner':7,'C-level':8,'CEO':9}
```

```python
x=num_col.drop(['Salary'],axis=1)
y=num_col.drop(['Level'],axis=1)

```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x,y)
```

```python
lr.score(x,y)
```

```python
#fitting the polynomial regression to the dataset
from sklearn.preprocessing import PolynomialFeatures
poly_regs=PolynomialFeatures(degree=4)
x_poly=poly_regs.fit_transform(x)
lr2=LinearRegression()
lr2.fit(x_poly,y)
```

```python
lr2.score(x_poly,y)
```

```python

#plotting of the linear regression
plt.scatter(x,y,color='blue')
plt.plot(x,lr.predict(x),color="red")
plt.title("bluff detection model(LinearRegression)")
plt.xlabel("Position levels")
plt.ylabel('Salary')
plt.show()
```

```python
#plotting of the ploynomial regression
plt.scatter(x,y,color='blue')
plt.plot(x,lr2.predict(poly_regs.fit_transform(x)),color='red')

plt.title("bluff detection model(polynomial regression)")

plt.xlabel('Position levels')
plt.ylabel('Salary')
plt.show()
```
