```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
tip=sns.load_dataset('tips')  #funciton is use to load the data set in program

print(tip.head())
```

```python
sns.catplot(x='day',y='total_bill',data=tip) #by default print the strip(dots)
sns.set()
```

```python
sns.catplot(x='day',y='total_bill',data=tip,kind='boxen')  #there are the multiple types of the kind in the seaborn
#interquartile range
sns.set()
```

```python
plt.figure(figsize=(9,9))  #figsize takes the length and breadth as the attribues (l,b)
# sns.countplot(tip['day'])
sns.boxenplot(tip['total_bill'])
tip.describe()
```

<h1> for sorting the values of the data set and save into the another data set

```python
sortvalues=tip.sort_values(by='total_bill')['total_bill']
print(sortvalues)
```

<h1> Dealing with the outliers using the seaborn <h1>

```python
#1.remove the outliers
#2. transforming and scaleing the outliers
#1
upper_limit=tip['total_bill'].quantile(0.9)
print(upper_limit)
```

```python
sns.boxplot(tip[tip['total_bill']<upper_limit]['total_bill'])
```

```python
#using iqr range
q1=np.percentile(tip['total_bill'],25) #it gives the 25 percentage of the tip['total_bill']
q3=np.percentile(tip['total_bill'],75)

iqr=q3-q1
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
print(iqr)
print(q1)
q3
```
