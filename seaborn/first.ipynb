```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data1=sns.load_dataset('tips')  #funciton is use to load the data set in program

print(data1.head())

```

```python
sns.lineplot(data1['total_bill'])
```

```python
tip=sns.load_dataset('tips')
sns.scatterplot(x='total_bill',y='tip',data=tip)
sns.set()  #for removing the upper line at the top of the graph

```

```python
sns.displot(tip['total_bill']) #it prints the bargaphs
sns.set()
```

```python
sns.distplot(tip['total_bill'])  #it prints the graphs with the line
#sns.distplot(tip['total_bill'],hist=False)  #it prints only the 
sns.set()
```
