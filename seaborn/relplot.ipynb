```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tip=sns.load_dataset('tips')  #funciton is use to load the data set in program

#there are two types of the plot 
# catplot()  --> categorical plot
# relplot()  --> relational plot
```

```python
sns.relplot(x='total_bill',y='tip',data=tip,kind='line')  #it prints the line plot

sns.relplot(x='total_bill',y='tip',data=tip) #it prints the dot plot
sns.set()
```

```python
sns.relplot(x='total_bill',y='tip',data=tip,hue='time')
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',style='smoker')
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',kind='line')
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',kind='line',style='smoker')
```

```python
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',style='smoker',col='time')
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',style='smoker',col='smoker')
```

```python
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',style='smoker',col='day')
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',kind='line',style='smoker',col='day')
#it divides the plots on the basis of the day

sns.set()
```

```python
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',style='smoker',row='day')
```

```python
sns.relplot(x='total_bill',y='tip',data=tip,hue='time',style='smoker',row='day',col='size')
```
